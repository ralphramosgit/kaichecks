# Frontend and API parity test

This notebook proves two things about the Kaichecks site:

1. The beach list is **live from the API**, not hardcoded. The site loads it from `GET /beaches`.
2. The safety rate shown on the dashboard is **computed from a real `POST /predict` call** plus each beach's real historical exceedance rate. Nothing is mocked, and there is no fake fallback.

It does this by calling the same backend the browser calls, then running the exact same math the frontend runs (a line-for-line mirror of `ui/src/lib/simulation.ts`). If the numbers printed here match the numbers on screen, the site is using real API data.

**How to use it**: start the backend, run all cells top to bottom, then open the site and compare the printed values to the dashboard badges. Each step below tells you exactly where to look.

## Setup

Prerequisites:

- The backend is running at `http://localhost:8000`. Start it with `docker compose up --build` from `backend/docker`, or with uvicorn from `backend/fastAPI`.
- The frontend dev server is running at `http://localhost:3000` (only needed for the visual comparison, not for this notebook to run).

This notebook uses only the Python standard library for HTTP (`urllib`) plus pandas for table display, so there is nothing extra to install.

```python
import json
import urllib.request
import urllib.error

# Base URL of the FastAPI backend. This is the same default the frontend uses
# (ui/src/lib/api.ts getApiBaseUrl, env NEXT_PUBLIC_API_BASE_URL).
BASE_URL = "http://localhost:8000"


# GET a JSON endpoint and return the parsed body.
def get(path):
    with urllib.request.urlopen(BASE_URL + path, timeout=10) as response:
        return json.load(response)


# POST a JSON body and return the parsed response.
def post(path, body):
    payload = json.dumps(body).encode("utf-8")
    request = urllib.request.Request(
        BASE_URL + path,
        data=payload,
        headers={"Content-Type": "application/json"},
        method="POST",
    )
    with urllib.request.urlopen(request, timeout=10) as response:
        return json.load(response)


print("Backend base URL:", BASE_URL)
health = get("/health")
print(
    "Health:",
    health["status"],
    "| models_loaded:",
    health["models_loaded"],
    "| primary_model:",
    health["primary_model"],
)
```

## Step 2: the exact safety math the frontend runs

The backend model is island-wide: one `POST /predict` returns a single `unsafe_probability` for the whole island. The frontend spreads that one probability across beaches using each beach's real historical exceedance rate as a relative-risk multiplier on the odds. This is the only transformation, and it uses no randomness.

The functions below are a line-for-line mirror of `ui/src/lib/simulation.ts` and `ui/src/lib/utils.ts`:

- `scale_probability_to_beach` mirrors `scaleProbabilityToBeach` (the odds-ratio scaling).
- `safety_from_probability` mirrors `safetyFromProbability` (unsafe at 0.5, caution at 0.25).
- `format_percent` mirrors `formatPercent` (how a badge renders a percentage).
- `run_simulation` mirrors `runSimulation` (per-beach probabilities plus the counts the badges show).

```python
import math

# Mirrors ui/src/lib/constants.ts SAFETY_THRESHOLDS.
THRESHOLD_UNSAFE = 0.5
THRESHOLD_CAUTION = 0.25

# Mirrors ui/src/lib/simulation.ts smoothing parameters.
# A beach needs roughly this many of its own samples before its raw rate
# outweighs the island average, so sparse beaches stop reading a hard 0%.
EXCEEDANCE_PRIOR_STRENGTH = 30
# Below this sample count a beach's history is flagged as limited data.
MIN_RELIABLE_SAMPLES = 30

# Mirrors ui/src/lib/constants.ts INTENSITY_PRESETS (mm per day, oldest first).
INTENSITY_PRESETS = {
    "dry": [0, 0, 0, 0, 0, 0, 0],
    "light": [1, 0, 2, 1, 0, 3, 2],
    "moderate": [4, 6, 3, 8, 5, 10, 7],
    "heavy": [10, 14, 9, 18, 22, 16, 24],
    "storm": [18, 26, 22, 30, 38, 44, 52],
}


# Mirror of smoothedExceedance in ui/src/lib/simulation.ts. Shrinks a beach's
# raw historical exceedance rate toward the island average (Laplace / empirical
# Bayes), so a beach with few samples regresses to the norm instead of claiming
# a confident 0% or spiking on a single unlucky sample.
def smoothed_exceedance(beach, island_mean_exceedance):
    exceedances = beach["samples"] * beach["exceedance_rate"]
    return (exceedances + EXCEEDANCE_PRIOR_STRENGTH * island_mean_exceedance) / (
        beach["samples"] + EXCEEDANCE_PRIOR_STRENGTH
    )


# Mirror of scaleProbabilityToBeach in ui/src/lib/simulation.ts. Scales the
# island-wide model probability to one beach using its (smoothed) exceedance
# rate as a relative-risk multiplier on the odds.
def scale_probability_to_beach(island_probability, beach_exceedance, mean_exceedance):
    p = min(1.0, max(0.0, island_probability))
    if p <= 0 or p >= 1 or mean_exceedance <= 0:
        return p
    relative_risk = beach_exceedance / mean_exceedance
    odds = (p / (1 - p)) * relative_risk
    return odds / (1 + odds)


# Mirror of safetyFromProbability in ui/src/lib/simulation.ts.
def safety_from_probability(probability):
    if probability >= THRESHOLD_UNSAFE:
        return "unsafe"
    if probability >= THRESHOLD_CAUTION:
        return "caution"
    return "safe"


# Mirror of formatPercent in ui/src/lib/utils.ts (JS toFixed rounds half up).
def format_percent(value, decimals=0):
    factor = 10 ** decimals
    rounded = math.floor(value * 100 * factor + 0.5) / factor
    return f"{rounded:.{decimals}f}%"


# Mirror of runSimulation in ui/src/lib/simulation.ts. The probability uses the
# smoothed exceedance, while limited_data flags beaches with few samples.
def run_simulation(island_probability, beach_records, mean_exceedance):
    rows = []
    for b in beach_records:
        smoothed = smoothed_exceedance(b, mean_exceedance)
        prob = scale_probability_to_beach(island_probability, smoothed, mean_exceedance)
        rows.append(
            {
                "location_id": b["location_id"],
                "location_name": b["location_name"],
                "samples": b["samples"],
                "exceedance_rate": b["exceedance_rate"],
                "smoothed_exceedance": smoothed,
                "unsafe_probability": prob,
                "safety_level": safety_from_probability(prob),
                "limited_data": b["samples"] < MIN_RELIABLE_SAMPLES,
            }
        )
    table = pd.DataFrame(rows)
    counts = table["safety_level"].value_counts()
    return {
        "predictions": table.sort_values(
            "unsafe_probability", ascending=False
        ).reset_index(drop=True),
        "unsafe_count": int(counts.get("unsafe", 0)),
        "caution_count": int(counts.get("caution", 0)),
        "safe_count": int(counts.get("safe", 0)),
        "average_probability": float(table["unsafe_probability"].mean()),
    }
```

## Step 6: prove there is no fake fallback (error path)

The site never invents numbers. If the model is unreachable it shows an error screen, not sample data.

To see it:

1. Stop the backend: `docker stop Kaichecks-api`.
2. On the site, change the storm. The app shows **Cannot reach the model** with no data.
3. Run the cell below. With the backend down it raises a `URLError` instead of returning a number, which is the same hard-failure behavior the frontend has.
4. Restart the backend (`docker compose up -d`) and click **Try again** on the site, or re-run this notebook. Real data returns.

```python
# Proof that there is no fake fallback. With the backend stopped
# (docker stop Kaichecks-api) this call raises instead of inventing numbers,
# exactly like the site, which shows "Cannot reach the model" and no data.
try:
    post("/predict", {"rainfall_7day": INTENSITY_PRESETS["moderate"], "month": 2})
    print("Backend reachable: the site shows live results.")
except urllib.error.URLError as error:
    print("Backend unreachable:", error)
    print("This is the error path. The frontend shows the error screen, never fake data.")
```
